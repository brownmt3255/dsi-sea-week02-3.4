### Intro: find missing data

```python
import pandas as pd
import numpy as np
import math

crunch = pd.read_csv('crunchbase_monthly_export.csv')
crunch.head()
```

```python
# Here's building a function using def: 

def square_root(x): return math.sqrt(x)
```

```python
# Here's building the same function using lambda

square_root = lambda x: math.sqrt(x)
square_root(100)
```

```python
# Use the isnull() method to detect the missing values. The output
# shows True when the value is missing. By adding an index into 
# the dataset, you obtain just the entries that are missing.
# A dataset could represent missing data in several ways. In this 
# example, you see missing data represented as np.NaN (NumPy Not 
# a Number) and the Python None value.

print crunch[crunch.isnull()]

#print crunch[crunch.isnull()].index.tolist()


```

### Intro: fill in missing data

```python
# To fill in missing data use fillna(). For fillna() you need to 
# provide a number. Usually, the mean, median, or mode is used. 
# Let's use the same data set and this time let's fill in missing 
# values with the mean. 

print crunch.fillna(int(crunch.mean()))
```

```python
# We could also just drop all the NAs, by using dropna()

crunch.dropna()
```

```python
# read in the raw data from the github url

crunch = pd.read_csv("https://raw.githubusercontent.com/suneel0101/lesson-plan/master/crunchbase_monthly_export.csv")
crunch.head()
```

```python
#Clean the funding_total_usd column (it's the wrong data type)
                                    
import re

crunch.rename(columns = {' funding_total_usd ': 'funding_total_usd'}, inplace = True)


def replaceShit(stringy):
   if stringy.__class__.__name__ == 'str':
       new_stringy = re.sub("[^0-9]", "",stringy)
       #For Alphabet things
       #new_stringy = re.sub("[^a-z]", "",stringy.lower())
       if not new_stringy:
           return 0
       return new_stringy
   return stringy

crunch['funding_total_usd'] = [float(replaceShit(x)) for x in crunch['funding_total_usd']]

print crunch['funding_total_usd'][0:20]
```

```python
# find missing values

# print crunch[crunch.isnull()]

sum(crunch['funding_total_usd'].isnull())
```

```python
# fill in missing data

crunch.category_list.fillna('Unkown')
crunch.homepage_url.fillna("")
crunch[' market '].fillna("Unknown")
crunch.status.fillna("Unknown")


```

```python
# create a pivot table on the data

pd.pivot_table(crunch,index=['country_code'," market "],values=["funding_total_usd","funding_rounds"],aggfunc=[np.sum],
               margins=True)
```
